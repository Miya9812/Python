```python
a=int(input("Saisir un nombre: "))
b=int(input("Saisir un autre nombre: "))
i=2
j=2
sa=0
sb=0
while i<=a/2:
    if a%i==0:
        sa+=i
    i+=1
while j<=b/2:
    if b%j==0:
        sb+=j
    j+=1
if sa==b and sb==a:
    print("Les deux nombres sont amis")
else:
    print("les deux nombres ne sont pas amis")
```
